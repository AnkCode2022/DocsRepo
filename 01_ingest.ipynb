```python
train_path = "/Volumes/ws1/default/v1/train.csv"
test_path  = "/Volumes/ws1/default/v1/test.csv"

train_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(train_path)
)

test_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(test_path)
)

train_df.write.format("delta").mode("overwrite").saveAsTable("ws1.default.bronze_train")
test_df.write.format("delta").mode("overwrite").saveAsTable("ws1.default.bronze_test")

print("Bronze tables created")
```
