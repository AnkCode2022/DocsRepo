```python
import mlflow.pyfunc
import pandas as pd

model = mlflow.pyfunc.load_model(
    "models:/ws1.default.vehicle_insurance_model/1"
)

df = spark.table("ws1.default.gold_test").toPandas()

ids = df["id"]
X = df.drop(columns=["id"])

pred = model.predict(X)

out = pd.DataFrame({
    "id": ids,
    "Response": pred
})

out.to_csv("/Volumes/ws1/default/v1/output/predictions.csv", index=False)

print("Predictions saved successfully")
```
